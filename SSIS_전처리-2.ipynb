```python
for i in range(0, len(my_real_removed_data.index)):
    for j in range(0, len(my_real_removed_data.columns)):
        if value_list[j] == "생년월일":
            if my_real_removed_data.loc[i, value_list[j]] >= 0 and my_real_removed_data.loc[i, value_list[j]] <= 3:
                my_real_removed_data.loc[i, value_list[j]] = 6
            elif my_real_removed_data.loc[i, value_list[j]] >= 4 and my_real_removed_data.loc[i, value_list[j]] <= 7:
                my_real_removed_data.loc[i, value_list[j]] = 5
            elif my_real_removed_data.loc[i, value_list[j]] >= 8 and my_real_removed_data.loc[i, value_list[j]] <= 13:
                my_real_removed_data.loc[i, value_list[j]] = 4
            elif my_real_removed_data.loc[i, value_list[j]] >= 14 and my_real_removed_data.loc[i, value_list[j]] <= 16:
                my_real_removed_data.loc[i, value_list[j]] = 3
            elif my_real_removed_data.loc[i, value_list[j]] >= 17 and my_real_removed_data.loc[i, value_list[j]] <= 19:
                my_real_removed_data.loc[i, value_list[j]] = 2
            else:
                my_real_removed_data.loc[i, value_list[j]] = 1

value_list = ['성별','생년월일','친권자 유형','가족 유형','다문화가족','기초생활수급 유형','접수 유형','신고접수 구분',
  '피해아동 상태 구분','행위자 아동 관계','아동 동거 여부','통계 거점','조사 완료 여부','아동 동거 여부.1','학대 혐의 여부']
```

```python
final_data = pd.DataFrame()

for j in range(0, len(my_real_removed_data.columns)):
    if value_list[j] == "친권자 유형":
        my_real_removed_data[value_list[j]] = my_real_removed_data[value_list[j]].map(parents)
    elif value_list[j] == "가족 유형":
        my_real_removed_data[value_list[j]] = my_real_removed_data[value_list[j]].map(family)
    elif value_list[j] == "다문화가족":
        my_real_removed_data[value_list[j]] = my_real_removed_data[value_list[j]].map(culture)
    elif value_list[j] == "기초생활수급 유형":
        my_real_removed_data[value_list[j]] = my_real_removed_data[value_list[j]].map(basic)
    elif value_list[j] == "접수 유형":
        my_real_removed_data[value_list[j]] = my_real_removed_data[value_list[j]].map(accepted)
    elif value_list[j] == "신고접수 구분":
        my_real_removed_data[value_list[j]] = my_real_removed_data[value_list[j]].map(call)
    elif value_list[j] == "피해아동 상태 구분":
        my_real_removed_data[value_list[j]] = my_real_removed_data[value_list[j]].map(child)
    elif value_list[j] == "행위자 아동 관계":
        my_real_removed_data[value_list[j]] = my_real_removed_data[value_list[j]].map(caller)
    elif value_list[j] == "아동 동거 여부":
        my_real_removed_data[value_list[j]] = my_real_removed_data[value_list[j]].map(livewith)
    elif value_list[j] == "통계 거점":
        my_real_removed_data[value_list[j]] = my_real_removed_data[value_list[j]].map(location)
    elif value_list[j] == "조사 완료 여부":
        my_real_removed_data[value_list[j]] = my_real_removed_data[value_list[j]].map(finish)
    elif value_list[j] == "아동 동거 여부.1":
        my_real_removed_data[value_list[j]] = my_real_removed_data[value_list[j]].map(livewith_1)
    elif value_list[j] == "학대 혐의 여부":
        my_real_removed_data[value_list[j]]= my_real_removed_data[value_list[j]].map(judge)

            
```
